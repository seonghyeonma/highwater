```python
# 자기회생여부 전처리
# 1. personal_rehabilitation_yn == 0 -> personal_rehabilitation_complete_yn = 0
# 2. personal_rehabilitation_complete_yn == 1 ->  personal_rehabilitation_yn == 1
# 3. personal_rehabilitation_yn == 0 인데 personal_rehabilitation_complete_yn == 1이면 제거 => 1번 전처리로 처리됨.

# 1
indexs_1 = df1[(df1['personal_rehabilitation_yn'] == 0 ) & (df1['personal_rehabilitation_complete_yn'] != 0)].index
df1.loc[indexs_1, ['personal_rehabilitation_complete_yn']] = 0
df1.isnull().sum()
```

```python
# 2
indexs_2 = df1[(df1['personal_rehabilitation_complete_yn'] == 1 ) & (df1['personal_rehabilitation_yn'] != 1)].index
indexs_2

df1.loc[indexs_2, ['personal_rehabilitation_yn']] = 1
df1.isnull().sum()
```
