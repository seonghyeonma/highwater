```python
# 필요 패키지 import
import pandas as pd
import numpy as np
```

```python
# 파일 업로드
user = pd.read_csv('user_spec_clean.csv')
loan = pd.read_csv('loan_result_clean.csv')
log = pd.read_csv('log_data_clean.csv')
```

```python
# user
# 1. birth_year, year, age_year -> age로 통일
# 2. insert_time -> 삭제?
# 3. company_enter_month, enter_month -> work_day로 통일
# 4. purpose -> 라벨 인코딩 
# 5. employment_type -> 라벨 인코딩
# 6. houseown_type -> 라벨 인코딩
# 7. income_type -> 라벨 인코딩
# 8. credit_score -> 범주
```

```python
# 1. birth_year, year -> age 있으니까 삭제?
user = user.drop('birth_year', axis = 1)
user = user.drop('year', axis = 1)
user = user.drop('age_year', axis = 1)
user
```

```python
# 2. insert_time -> 삭제?
user = user.drop('insert_time', axis = 1)
user
```

```python
# 3. company_enter_month, enter_month -> work_day 통일
user = user.drop('company_enter_month', axis = 1)
user = user.drop('enter_month', axis = 1)
user 
```

```python
# Label encoding 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le
```

```python
# 4. purpose -> 라벨 인코딩 
label_purpose = user['purpose']
print(label_purpose.unique())

le.fit(label_purpose)
label_encoded = le.transform(label_purpose)
label_encoded_df = pd.DataFrame(label_encoded, columns = ['label_purpose'])

user = pd.concat([user, label_encoded_df],axis = 1)
user
```

```python
# 5. employment_type -> 라벨 인코딩
label_employment_type = user['employment_type']
print(label_employment_type.unique())

le.fit(label_employment_type)
label_encoded = le.transform(label_employment_type)
label_encoded_df = pd.DataFrame(label_encoded, columns = ['label_employment_type'])

user = pd.concat([user, label_encoded_df], axis = 1)
user
```

```python
# 6. houseown_type -> 라벨 인코딩
label_houseown_type = user['houseown_type']
print(label_houseown_type.unique())

le.fit(label_houseown_type)
label_encoded = le.transform(label_houseown_type)
label_encoded_df = pd.DataFrame(label_encoded, columns = ['label_houseown_type'])

user = pd.concat([user, label_encoded_df], axis = 1)
user
```

```python
# 7. income_type -> 라벨 인코딩
label_income_type = user['income_type']
print(label_income_type.unique())

le.fit(label_income_type)
label_encoded = le.transform(label_income_type)
label_encoded_df = pd.DataFrame(label_encoded, columns = ['label_income_type'])

user = pd.concat([user, label_encoded_df], axis = 1)
user
```
