```python
def find_outlier(data):
    q1, q3 = np.percentile(data, [25, 75])
    
    iqr = q3 - q1
    
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    mask = np.where((data > upper_bound) | (data < lower_bound))
    # mask는 튜플 형태이므로 튜플을 제거한다.
    return mask[0].tolist()
```

```python
# loan_limit / yearly_income / desired_amount / existing_loan_amt
loan_limit_out = find_outlier(train['loan_limit'])
yearly_income_out = find_outlier(train['yearly_income'])
desired_amount_out = find_outlier(train['desired_amount'])
existing_loan_amt_out = find_outlier(train['existing_loan_amt'])
out = loan_limit_out + yearly_income_out + desired_amount_out + existing_loan_amt_out 
```
